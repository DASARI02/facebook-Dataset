```python
#import pandas
import pandas as pd
```

```python
data=pd.read_excel('facebook Dataset.xlsx')
data.tail()
```

Data cleaning

```python
data['GENDER'] = data['GENDER'].str.replace('FEMALE', 'Female')
data.tail()
```

```python
data.describe()
```

**Mean**

```python
#average of no of likes received and  average no of posts
n = len(data['USER NAME'])
sum1=0
sum2=0
for i,j in zip(data['NO OF LIKES  RECEIVED'],data['NO Of POSTS']):
    sum1+=i
    sum2+=j
print('average of no of likes received: ',sum1/n)
print()
print('average no of posts: ',sum2/n)
```

**Median**

```python
#median of age
s = sorted(data['AGE'])
if  n% 2 == 0:
    median1 = s[n//2]
    median2 = s[n//2 - 1]
    median = (median1 + median2)/2
else:
    median = s[n//2]
print('median of age: ',median)
```

**Mode**

```python
#max of friends

m=data['NO OF FRIENDS'][0]
realfriendno =  data['NO OF FRIENDS']<=5000
df = data[realfriendno]

for i in df['NO OF FRIENDS']:
    if(i>m):
        m=i

df2=df.loc[df['NO OF FRIENDS'] == m, 'USER NAME'].iloc[0]
print('the user with more number of friends: ',df2,', with number of friends: ',m)
```
